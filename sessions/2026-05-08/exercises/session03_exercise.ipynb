## Part D: hidden state demo

This part demonstrates the hidden state problem in notebooks.

**D1.** Run the cell below.

````python
secret_variable = "I was set in cell D1"
````

**D2.** Now delete the cell above (select it and press `D D` in Positron, or use the delete button).

**D3.** Run the cell below. It still works -- even though the cell that defined `secret_variable` no longer exists.

````python
print(secret_variable)
````

**D4.** Now go to **Kernel -> Restart Kernel** and run the cell above again. What happens?

This is the hidden state problem. The notebook appeared to work, but only because of something that no longer exists in the file. `Restart & Run All` is the only reliable check.
