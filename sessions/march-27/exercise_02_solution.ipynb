# Functions

```python
def normalise(x):
    mn = min(x)
    mx = max(x)
    return [(v - mn) / (mx - mn) for v in x]

normalised_scores = normalise(scores)
print(f"Normalised: {[round(v, 3) for v in normalised_scores]}")

```
